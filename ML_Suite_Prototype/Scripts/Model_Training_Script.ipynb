```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('../Data/combo_df')
```

```python
df.head()
```

```python
df.drop(['ID','ID.1'],axis=1,inplace=True)
```

```python
df.set_index('combo_index',inplace=True)
```

```python
inputs=df.drop('Complementary', axis=1)
output=df['Complementary']
```

```python
objects=[i for i in inputs.columns if type(inputs[i][1]) is str]
numerical=[i for i in inputs.columns if type(inputs[i][1]) is np.float64]
```

```python
dummy_df=pd.get_dummies(df[objects])
dummy_df.head()
```

```python
inputs.drop(objects,axis=1,inplace=True)
```

```python
X=pd.concat([inputs,dummy_df],axis=1)
```

```python
encoder_inst=LabelEncoder()
y=encoder_inst.fit_transform(output)
```

```python
len(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25)
```

```python
rf_inst=RandomForestClassifier()
rf_fit=rf_inst.fit(X_train,y_train)
rf_pred=rf_fit.predict(X_test)
accuracy_score(rf_pred,y_test)
```
